Setelah kita mendefinisikan fungsi, kita dapat menggunakannya di dalam program kita dengan memberikan argumen yang sesuai. Dalam contoh di atas, kita menggunakan fungsi `hitung_luas_persegi` untuk menghitung luas persegi dengan sisi 5.

Ini adalah contoh sederhana dari bagaimana mahasiswa dapat membuat dan menggunakan fungsi dalam Python untuk menyelesaikan berbagai masalah pemrograman.

```python
import numpy as np
import matplotlib.pyplot as plt

# Nilai-nilai yang diberikan
A = 2  # Amplitude (m)
k = 4  # Bilangan gelombang (m^-1)
omega = 3 * np.pi  # Kecepatan sudut (rad/s)
t = 2  # Waktu (s)

# Array x dari 0 hingga 5 m
x = np.linspace(0, 5, 100)

# Menghitung simpangan gelombang pada t = 2 s
y = A * np.sin(k * x - omega * t)

# Plotting simpangan gelombang
plt.plot(x, y)
plt.title('Simpangan Gelombang pada t = 2 s')
plt.xlabel('x (m)')
plt.ylabel('Simpangan (m)')
plt.grid(True)
plt.show()
```

Persamaan gelombang yang diberikan dapat digunakan untuk memvisualisasikan gelombang berjalan pada waktu tertentu. Persamaan ini adalah $y(x,t)=A sin(kxt)$. Dalam hal ini, gelombang ini digambarkan sebagai gelombang sinusoidal yang merambat dalam satu dimensi dengan simpangan y pada posisi x dan waktu t. Dengan menggunakan nilai-nilai yang diberikan dalam kode yang disajikan, seperti amplitudo A, bilangan gelombang k, dan kecepatan sudut , simpangan gelombang dihitung pada waktu tertentu, yaitu t = 2 s. Proses ini dilakukan dengan menggunakan persamaan gelombang untuk setiap nilai posisi x dalam rentang dari 0 hingga 5 meter. Hasilnya, yang merupakan simpangan gelombang pada titik-titik  tertentu pada waktu  = 2 s, kemudian diplot dalam grafik menggunakan modul `matplotlib.pyplot`. Grafik yang dihasilkan memberikan gambaran visual bentuk gelombang sinusoidal pada waktu yang ditentukan. Ini memungkinkan kita untuk melihat bagaimana gelombang berperilaku dalam medium pada waktu dan tempat tertentu. Ini dapat sangat membantu dalam pemodelan dan pemahaman fenomena gelombang dalam berbagai konteks fisika.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan parameter gelombang
A = 2  # Amplitudo
k = 4  # Bilangan gelombang
omega = 3 * np.pi  # Kecepatan sudut
phi = np.pi / 2  # Beda fase

# Array posisi x dari 0 hingga 5 meter
x = np.linspace(0, 5, 500)

# Fungsi gelombang y1
def y1(x, t):
    return A * np.sin(k * x - omega * t)

# Fungsi gelombang y2
def y2(x, t):
    return A * np.sin(k * x - omega * t + phi)

# Fungsi gelombang y3 (penjumlahan y1 dan y2)
def y3(x, t):
    return y1(x, t) + y2(x, t)

# Waktu t = 2 s
t = 2

# Plotting
plt.figure(figsize=(10, 6))

# Subplot untuk gelombang y1
plt.subplot(3, 1, 1)
plt.plot(x, y1(x, t), label='y1')
plt.xlabel('x (m)')
plt.ylabel('Simpangan')
plt.title('Gelombang y1')
plt.legend()

# Subplot untuk gelombang y2
plt.subplot(3, 1, 2)
plt.plot(x, y2(x, t), label='y2')
plt.xlabel('x (m)')
plt.ylabel('Simpangan')
plt.title('Gelombang y2')
plt.legend()

# Subplot untuk gelombang y3
plt.subplot(3, 1, 3)
plt.plot(x, y3(x, t), label='y3')
plt.xlabel('x (m)')
plt.ylabel('Simpangan')
plt.title('Gelombang y3 = y1 + y2')
plt.legend()

plt.tight_layout()
plt.show()
```

Program Python di atas digunakan untuk menghasilkan tiga plot dalam satu grafik yang menampilkan gelombang bergerak dalam satu dimensi pada  = 2 s. Gelombang-gelombang ini dihitung menggunakan fungsi gelombang yang sesuai, yaitu `y1`, `y2`, dan `y3`.

Pertama, program menentukan amplitudo (A), bilangan gelombang (k), kecepatan sudut (), dan beda fase (). Selanjutnya, array `x` dibuat untuk menyimpan nilai posisi x antara 0 dan 5 meter.

Selanjutnya, fungsi gelombang `y1`, `y2`, dan `y3` didefinisikan sesuai dengan persamaan gelombang yang diberikan. Fungsi `y1` dan `y2` menunjukkan dua gelombang bergerak yang berbeda, dan fungsi "y3" menunjukkan gabungan dari kedua gelombang.

Kemudian, pada waktu  = 2 s, nilai simpangan dari ketiga gelombang dihitung menggunakan fungsi gelombang yang sesuai. Hasil perhitungan tersebut kemudian digunakan untuk membuat plot menggunakan modul `matplotlib.pyplot`. 

Untuk membuat tiga subplot dalam satu grafik vertikal, program menggunakan `plt.subplot`. Setiap subplot menampilkan gelombang dengan label sumbu dan judul yang sesuai. Kemudian, grafik yang dihasilkan ditampilkan menggunakan `plt.show()`.

Program ini memungkinkan pengguna untuk melihat bagaimana gelombang bergerak dalam satu dimensi dan berinteraksi satu sama lain pada waktu $t = 2 s$.
