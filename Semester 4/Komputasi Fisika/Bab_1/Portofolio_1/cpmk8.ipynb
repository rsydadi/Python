```python
# Contoh
import numpy as np
import matplotlib.pyplot as plt

plt.plot([16, 2, 3, 9, 3, 2, 16], color='b', marker='H', linewidth=2, markersize=12)
plt.show()
```

```python
# Membuat array nilai x dari 0 sampai 4 dengan langkah 0.1
x = np.arange(0, 4*np.pi, 0.1)

# Membuat subplot dengan ukuran 2x2
plt.figure(figsize=(10, 8))

# Subplot 1: sin(x)
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x), 'r')
plt.title('sin(x)')

# Subplot 2: cos(x)
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), 'g')
plt.title('cos(x)')

# Subplot 3: sec(x)
plt.subplot(2, 2, 3)
sec_x = 1 / np.cos(x)
plt.plot(x, sec_x, 'b')
plt.title('sec(x)')

# Subplot 4: cosec(x)
plt.subplot(2, 2, 4)
cosec_x = 1 / np.sin(x)
plt.plot(x, cosec_x, 'm')
plt.title('cosec(x)')

plt.tight_layout() # Untuk memastikan subplot tidak tumpang tindih
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
def pmgauss(x, y): 
 r1 = (x-1)**2 + (y-2)**2 
 r2 = (x-3)**2 + (y-1)**2 
 return 2*np.exp(-0.5*r1) - 3*np.exp(-2*r2) 
a, b = 4, 3 
x = np.linspace(0, a, 60) 
y = np.linspace(0, b, 45) 
X, Y = np.meshgrid(x, y) 
Z = pmgauss(X, Y) 
fig, ax = plt.subplots(1, 2, figsize=(9.2, 4), 
subplot_kw={'projection': '3d'}) 
for i in range(2): 
 ax[i].set_zlim(-3, 2) 
 ax[i].xaxis.set_ticks(range(a+1)) # manually set ticks 
 ax[i].yaxis.set_ticks(range(b+1)) 
 ax[i].set_xlabel(r'$x$') 
 ax[i].set_ylabel(r'$y$') 
 ax[i].set_zlabel(r'$f(x,y)$') 
 ax[i].view_init(40, -30) 
 
# Plot wireframe dan permukaan 
plt.subplots_adjust(left=0.04, bottom=0.04, right=0.96, top=0.96, wspace=0.05)
p0 = ax[0].plot_wireframe(X, Y, Z, rcount=40, ccount=40, color='C1') 
p1 = ax[1].plot_surface(X, Y, Z, rcount=50, ccount=50, color='C1') 
plt.subplots_adjust(left=0.0) 
plt.show()
```
