```python
import numpy as np
import matplotlib.pyplot as plt

# Diketahui
v_awal = 2.4  # kecepatan awal in m/s
g = 10  # percepatan gravitasi
h = 12  # ketinggian benda

# waktu yang diperlukan untuk mencapai ketinggian maksimum
t = np.arange(0, 5.1, 0.1)

# analitis
v_analitis = v_awal + g*t

# numerik
v_numerik = np.zeros_like(t)
v_numerik[0] = v_awal
for i in range(1, len(t)):
    v_numerik[i] = v_numerik[i-1] + g*(t[i] - t[i-1])

# plot
plt.figure(figsize=(10, 7))
plt.plot(t, v_analitis, label='Analitis')
plt.plot(t, v_numerik, 'o', label='Numerik')
plt.xlabel('Time (s)')
plt.ylabel('Kecepatan (m/s)')
plt.title('Kecepatan vs Waktu untuk Benda yang Dilempar ke Atas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

def f(x):
    return 2*x + 5

def integral_numerik(f, a, b, N):
    dx = (b - a) / N
    x = np.linspace(a, b, N+1)
    y = f(x)
    hasil = dx * (0.5*y[0] + 0.5*y[-1] + np.sum(y[1:-1]))
    return hasil

a = 2
b = 4
N_values = [10, 100, 1000, 10000, 100000]

for N in N_values:
    hasil = integral_numerik(f, a, b, N)
    print(f'Hasil dengan N={N}: {hasil}')

# Hasil analitik dari integral ini adalah 18
print('Hasil analitik: 18')
```
