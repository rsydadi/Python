```python
import numpy as np
def turunan(f, x, h): 
    numerator = f(x + h) - f(x - h) 
    derivatif = numerator / ( 2.0 * h ) 
    return derivatif

h = 0.01
x = 0
f = np.exp
derivatif = turunan(f, x, h)
print("Aproksimasi turunan exp(x) pada x = 0 adalah: %f. Errornya ada;ah %f."
      %(derivatif, abs(derivatif - 1)))

x = 0
def g(x):
    return np.exp(-2*x)
derivatif = turunan (g, x, h)
print("Aproksimasi tueunan exp(-2*x)pada x = adalah: {0:.5f}. Errornya adalah {1:.5f}."
     .format(derivatif, abs(derivatif-(-2.0))))

x = 2*np.pi
f = np.cos 
derivatif = turunan(f, x, h) 
print("Aproksimasi turunan cos(x) pada x = 2*pi adalah: {0:.5f}.  Errornya adalah {1:.5f}." 
      .format(derivatif, abs(derivatif - 0)))

x = 1 
f = np.log 
derivatif = turunan(f, x, h) 
print("Aproksimasi turunan ln(x) pada x = 0 adalah: {0:.5f}.  Errornya adalah {1:.5f}." 
      .format(derivatif, abs(derivatif - 1)))

#Pengaruh ukuran grid terhadap akurasi
#Metode beda maju
def beda_maju(f, x, h):
    fx = f(x) 
    fxph = f(x + h) 
    return (fxph - fx) / h 

#Metode beda mundur
def beda_mundur(f, x, dx): 
    fx = f(x) 
    fxnh = f(x - dx) 
    return (fx - fxnh) / dx 

#Metode beda pusat
def beda_pusat(f, x, h): 
    fxph = f(x + h) 
    fxnh = f(x - h) 
    return (fxph - fxnh) / (2 * h) 

eksak = np.cos(0.8)  # turunan dari sin(x) adalah cos(x) 
print("Turunan eksak sin(0.8) = %.5f." % eksak)
```

Kode tersebut bertujuan untuk mengaproksimasi turunan dari beberapa fungsi matematika menggunakan metode beda pusat dan mengevaluasi error yang dihasilkan. Kode pertama mendefinisikan fungsi `turunan` yang menerima fungsi `f`, titik `x`, dan langkah `h`. Fungsi ini menghitung turunan menggunakan formula beda pusat $\frac{f(x+h) - f(x-h)}{2h}$. Kemudian, turunan dihitung untuk beberapa fungsi pada titik tertentu: $e^x$ pada $x = 0$, $e^{-2x}$ pada $x = 0$, $\cos(x)$ pada $x = 2\pi$, dan $\ln(x)$ pada $x = 1$. Error dihitung sebagai selisih antara hasil aproksimasi dan nilai eksak dari turunan.

Selanjutnya, kode mendefinisikan tiga fungsi untuk menghitung turunan menggunakan metode beda maju, beda mundur, dan beda pusat. Fungsi `beda_maju` menghitung turunan dengan formula beda maju \(\frac{f(x+h) - f(x)}{h}\). Fungsi `beda_mundur` menghitung turunan dengan formula beda mundur $\frac{f(x) - f(x-h)}{h}$. Fungsi `beda_pusat` menggunakan formula beda pusat yang sama dengan fungsi `turunan`.

Akhirnya, kode mengevaluasi pengaruh ukuran grid terhadap akurasi metode dengan menghitung turunan dari $\sin(x)$ pada $x = 0.8$ menggunakan nilai eksak dari turunan yaitu $\cos(0.8)$ dan menampilkan hasilnya.
```python
import numpy as np
import matplotlib.pyplot as plt
# Metode beda maju
def beda_maju(f, x, h):
    return (f(x + h) - f(x)) / h

# Metode beda mundur
def beda_mundur(f, x, h):
    return (f(x) - f(x - h)) / h

# Metode beda pusat
def beda_pusat(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

# Fungsi eksak
def eksak(x):
    return np.cos(x)
            
# Header untuk output error
print('-' * 100)
print('%10s %20s %25s %30s' % ('h', 'Beda maju', 'Beda mundur', 'Beda pusat')) 
print('-' * 100)

# Membuat list kosong untuk menyimpan error untuk masing-masing metode
error_beda_maju = []
error_beda_mundur = []
error_beda_pusat = []
h_total = []

h = 1.0  # h awal
for i in range(10):
    bm = beda_maju(np.sin, 0.8, h)
    bd = beda_mundur(np.sin, 0.8, h)
    bp = beda_pusat(np.sin, 0.8, h)
    
    print('%10g %20g (error=%10.2g) %25g (error=%10.2g) %30g (error=%10.2g)'
          % (h, bm, abs(bm - eksak(0.8)), bd, abs(bd - eksak(0.8)), bp, abs(bp - eksak(0.8))))
    
    # Menyimpan h dan data error masing-masing
    h_total.append(h)
    error_beda_maju.append(abs(bm - eksak(0.8)))
    error_beda_mundur.append(abs(bd - eksak(0.8)))
    error_beda_pusat.append(abs(bp - eksak(0.8)))
    
    h = h / 2  # h berkurang setelahnya untuk iterasi selanjutnya

# Plot font
plt.rcParams['font.family'] = 'Times New Roman'

# Plotting
fig = plt.figure(figsize=(7, 5))
ax1 = plt.subplot(111)

ax1.loglog(h_total, error_beda_maju, 'b.-', label='Beda maju')
ax1.loglog(h_total, error_beda_mundur, 'r.-', label='Beda mundur')
ax1.loglog(h_total, error_beda_pusat, 'k.-', label='Beda pusat')

ax1.set_xlabel('$h$', fontsize=16)
ax1.set_ylabel('Error', fontsize=16)
ax1.set_title('Plot konvergensi', fontsize=16)
ax1.grid(True)
ax1.legend(loc='best', fontsize=14)

plt.show()
```

Pertama, tiga fungsi didefinisikan untuk menghitung turunan dengan metode beda maju, beda mundur, dan beda pusat.
```python
# Nomor 5

import numpy as np
import matplotlib.pyplot as plt

mu = 0.01  # kg/m
T0 = 100  # N
beta = 0.05  # konstanta
x = np.linspace(0, 10, 100)  # dari 0 hingga 10 meter
T = T0 * (1 + beta * x)
v_eksak = np.sqrt(T / mu)
dx = x[1] - x[0]

# Solusi eksak untuk dv/dx
dv_dx_eksak = np.gradient(v_eksak, dx)

# Pendekatan beda maju
dv_dx_maju = np.zeros_like(v_eksak)
dv_dx_maju[:-1] = (v_eksak[1:] - v_eksak[:-1]) / dx
dv_dx_maju[-1] = (v_eksak[-1] - v_eksak[-2]) / dx  # Asumsi untuk titik terakhir

# Pendekatan beda mundur
dv_dx_mundur = np.zeros_like(v_eksak)
dv_dx_mundur[1:] = (v_eksak[1:] - v_eksak[:-1]) / dx
dv_dx_mundur[0] = (v_eksak[1] - v_eksak[0]) / dx  # Asumsi untuk titik pertama

# Pendekatan beda pusat
dv_dx_pusat = np.zeros_like(v_eksak)
dv_dx_pusat[1:-1] = (v_eksak[2:] - v_eksak[:-2]) / (2 * dx)
dv_dx_pusat[0] = (v_eksak[1] - v_eksak[0]) / dx  # Asumsi untuk titik pertama
dv_dx_pusat[-1] = (v_eksak[-1] - v_eksak[-2]) / dx  # Asumsi untuk titik terakhir

# Plot hasil
plt.figure(figsize=(12, 8))
plt.plot(x, dv_dx_eksak, label='Solusi Eksak', color='blue')
plt.plot(x, dv_dx_maju, label='Pendekatan Beda Maju', linestyle='--', color='red')
plt.plot(x, dv_dx_mundur, label='Pendekatan Beda Mundur', linestyle='--', color='green')
plt.plot(x, dv_dx_pusat, label='Pendekatan Beda Pusat', linestyle='--', color='purple')
plt.xlabel('Posisi x (m)')
plt.ylabel('Gradien dv/dx')
plt.title('Gradien Kecepatan Gelombang Sepanjang Tali')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Nomor 1a

import numpy as np
import matplotlib.pyplot as plt

# Fungsi F(x)
def F(x):
    return 5 * x**2

# Solusi analitis
solusi_analitis = (5 / 3) * (2**3 - 0**3)

# Aturan Titik Tengah
def aturan_titik_tengah(f, a, b, n):
    h = (b - a) / n
    titik_tengah = a + h * (np.arange(n) + 0.5)
    return h * np.sum(f(titik_tengah))

# Aturan Trapesium
def aturan_trapesium(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    return (h / 2) * (f(x[0]) + 2 * np.sum(f(x[1:n])) + f(x[n]))

# Aturan Simpson
def aturan_simpson(f, a, b, n):
    if n % 2 == 1:
        n += 1  # Aturan Simpson memerlukan jumlah interval genap
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    return (h / 3) * (f(x[0]) + 4 * np.sum(f(x[1:n:2])) + 2 * np.sum(f(x[2:n-1:2])) + f(x[n]))

# Jumlah interval yang akan digunakan
jumlah_interval = [10, 20, 50, 100, 200]

# Hitung error untuk setiap metode dan jumlah interval
error_titik_tengah = []
error_trapesium = []
error_simpson = []

for n in jumlah_interval:
    hasil_titik_tengah = aturan_titik_tengah(F, 0, 2, n)
    hasil_trapesium = aturan_trapesium(F, 0, 2, n)
    hasil_simpson = aturan_simpson(F, 0, 2, n)
    
    error_tt = abs(hasil_titik_tengah - solusi_analitis)
    error_trap = abs(hasil_trapesium - solusi_analitis)
    error_simp = abs(hasil_simpson - solusi_analitis)
    
    error_titik_tengah.append(error_tt)
    error_trapesium.append(error_trap)
    error_simpson.append(error_simp)

# Tampilkan hasil dalam format tabel
print(f"{'Jumlah Interval':<15} {'Error Titik Tengah':<20} {'Error Trapesium':<20} {'Error Simpson':<20}")
print('-' * 75)
for i in range(len(jumlah_interval)):
    print(f"{jumlah_interval[i]:<15} {error_titik_tengah[i]:<20.10e} {error_trapesium[i]:<20.10e} {error_simpson[i]:<20.10e}")
```

Hasil diperoleh dari program yang menggunakan aturan titik tengah, trapezium, dan Simpson untuk menghitung integral numerik dari fungsi $F(x) = 5x^2$ pada interval 0,2. Untuk menentukan interval yang digunakan dalam integrasi numerik, program tersebut memiliki daftar "interval".
```python
# Nomor 1b

import numpy as np
import matplotlib.pyplot as plt

# Fungsi F(x)
def F(x):
    return 5 * x**2

# Solusi analitis untuk integral
solusi_analitis = lambda x: (5 / 3) * x**3

# Aturan Titik Tengah
def aturan_titik_tengah(f, a, b, n):
    h = (b - a) / n
    titik_tengah = a + h * (np.arange(n) + 0.5)
    return h * np.sum(f(titik_tengah))

# Aturan Trapesium
def aturan_trapesium(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    return (h / 2) * (f(x[0]) + 2 * np.sum(f(x[1:n])) + f(x[n]))

# Aturan Simpson
def aturan_simpson(f, a, b, n):
    if n % 2 == 1:
        n += 1 
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    return (h / 3) * (f(x[0]) + 4 * np.sum(f(x[1:n:2])) + 2 * np.sum(f(x[2:n-1:2])) + f(x[n]))

# Jumlah interval
n = 10
a, b = 0, 2
h = (b - a) / n
x = np.linspace(a, b, n + 1)

# Hasil numerik untuk setiap metode
titik_tengah = a + h * (np.arange(n) + 0.5)
hasil_titik_tengah = [aturan_titik_tengah(F, 0, xi, n) for xi in x]
hasil_trapesium = [aturan_trapesium(F, 0, xi, n) for xi in x]
hasil_simpson = [aturan_simpson(F, 0, xi, n) for xi in x]

# Solusi analitis
hasil_analitis = solusi_analitis(x)

# Plot hasilnya
plt.figure(figsize=(10, 6))
plt.plot(x, hasil_analitis, label='Analitis', color='black')
plt.plot(x, hasil_titik_tengah, label='Aturan Titik Tengah', marker='o', linestyle='--')
plt.plot(x, hasil_trapesium, label='Aturan Trapesium', marker='s', linestyle='-.')
plt.plot(x, hasil_simpson, label='Aturan Simpson', marker='^', linestyle=':')

plt.xlabel('Posisi (x)')
plt.ylabel('Usaha total (W)')
plt.title('Perbandingan Metode Integrasi Numerik')
plt.legend()
plt.grid()
plt.show()
```

Program ini menggunakan tiga metode numerik: Aturan Titik Tengah, Aturan Trapesium, dan Aturan Simpson, untuk menghitung integral numerik dari fungsi $F(x) = 5x^2$ pada interval 0,2. Setiap metode diimplementasikan dalam fungsi tersendiri (`midpoint_rule`, `trapezoidal_rule`, dan `simpson_rule`). Kemudian, program menghitung hasil numerik dari setiap metode pada serangkaian titik $(x)$ yang terpisah secara merata pada interval tersebut. Solusi analitis juga dihitung menggunakan fungsi analitis yang didefinisikan sebelumnya. Selanjutnya, hasil numerik dari setiap metode dan solusi analitis diplot untuk membandingkan kinerja dan akurasi metode numerik tersebut.
