```python
# Nomor 1a
# Fungsi menghitung polinomial Lagrange
import numpy as np
def lagrange(x, y, xi):
    n = len(x)
    yi = 0
    for i in range(n):
        # Menghitung basis Lagrange
        L = 1
        for j in range(n):
            if j != i:
                L *= (xi - x[j]) / (x[i] - x[j])
        # Menambahkan titik data ke interpolasi
        yi += y[i] * L
    return yi
# Data waktu dan temperatur Gunung Bromo
x = np.array([6, 9, 12, 15, 18, 20, 21, 24])
y = np.array([15, 18, 25, 22, 20, 19, 17, 16])
# Titik untuk interpolasi
xi = np.arange(6, 25, 1)
# Interpolasi dengan polinomial Lagrange
yi = lagrange(x, y, xi)
# Menampilkan hasil interpolasi
print("Hasil Interpolasi Temperatur:")
for jam, temperatur in zip(xi, yi):
    print(f"Jam: {jam}, Temperatur: {temperatur:.2f} °C")
```

```python
# Nomor 1b
import numpy as np
import matplotlib.pyplot as plt

# Fungsi menghitung polinomial Lagrange
def lagrange(x, y, xi):
    n = len(x)
    yi = np.zeros_like(xi)
    for i in range(n):
        # Menghitung basis Lagrange
        L = np.ones_like(xi)
        for j in range(n):
            if j != i:
                L *= (xi - x[j]) / (x[i] - x[j])
        # Menambahkan titik data ke interpolasi
        yi += y[i] * L
    return yi

# Data waktu dan temperatur Gunung Bromo
x = np.array([6, 9, 12, 15, 18, 20, 21, 24])
y = np.array([15, 18, 25, 22, 20, 19, 17, 16])

# Titik waktu untuk interpolasi
xi = np.arange(6, 25, 0.1)

# Interpolasi linier
yi_linier = np.interp(xi, x, y)

# Interpolasi polinomial Lagrange
yi_lagrange = lagrange(x, y, xi)

# Plot data mentah, hasil interpolasi linier, dan interpolasi Lagrange
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'go', label='Data Mentah')  # Data mentah berwarna hijau
plt.plot(xi, yi_linier, 'b--', label='Interpolasi Linier')  # Garis interpolasi linier berwarna biru
plt.plot(xi, yi_lagrange, 'r-', label='Interpolasi Lagrange')  # Garis interpolasi Lagrange berwarna merah
plt.title('Interpolasi Temperatur Gunung Bromo')
plt.xlabel('Jam')
plt.ylabel('Temperatur (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

Fungsi ini memiliki tiga parameter: $x$ untuk array dari titik waktu, $y$ untuk array dari nilai suhu yang sesuai dengan waktu, dan $xi$ untuk array dari titik waktu yang ingin diinterpolasi. Dimana $i$ adalah indeks titik data $j$ adalah indeks iterasi untuk membuat basis Lagrange, $xi$ adalah titik waktu pada indeks $i$ dan $xj$ adalah titik waktu pada indeks $j$ yang tidak sama dengan $i$. Setelah basis Lagrange dihitung maka nilai interpolasi pada titik waktu $xi$ dihitung dengan menjumlahkan hasil perkalian setiap nilai temperatur dengan basis Lagrange yang sesuai.
```python
# Nomor 3a

import numpy as np
import matplotlib.pyplot as plt

# Fungsi asli
def original_function(x):
    return x**3 - 2*x**2 + 2*x - 1

# Titik sampel
x_samples = np.array([1, 2, 3, 4])
y_samples = original_function(x_samples)
# Polinomial Lagrange L1
def lagrange_L1(x):
    return (x - 3)/(2 - 3) * original_function(2) + (x - 2)/(3 - 2) * original_function(3)

# Polinomial Lagrange L2
def lagrange_L2(x):
    return ((x - 3)*(x - 4))/((1 - 3)*(1 - 4)) * original_function(1) + ((x - 1)*(x - 4))/((3 - 1)*(3 - 4)) * original_function(3) + ((x - 1)*(x - 3))/((4 - 1)*(4 - 3)) * original_function(4)

# Polinomial Lagrange L4
def lagrange_L4(x):
    n = len(x_samples)
    L = 0
    for i in range(n):
        term = 1
        for j in range(n):
            if j != i:
                term *= (x - x_samples[j]) / (x_samples[i] - x_samples[j])
        L += y_samples[i] * term
    return L
# Plotting
x_values = np.linspace(0.5, 4.5, 400)
y_original = original_function(x_values)
y_L1 = lagrange_L1(x_values)
y_L2 = lagrange_L2(x_values)
y_L4 = lagrange_L4(x_values)

plt.figure(figsize=(18, 6))

# Plot fungsi asli dan interpolasi L1
plt.subplot(1, 3, 1)
plt.plot(x_values, y_original, label='Fungsi Asli')
plt.plot(x_values, y_L1, label='L1 (Linier)')
plt.scatter(x_samples[1:3], y_samples[1:3], color='red', label='Titik Sampel')
plt.title('Interpolasi Polinomial Lagrange L1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Plot fungsi asli dan interpolasi L2
plt.subplot(1, 3, 2)
plt.plot(x_values, y_original, label='Fungsi Asli')
plt.plot(x_values, y_L2, label='L2 (Kuadratik)')
plt.scatter(x_samples[[0, 2, 3]], y_samples[[0, 2, 3]], color='red', label='Titik Sampel')
plt.title('Interpolasi Polinomial Lagrange L2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Plot fungsi asli dan interpolasi L4
plt.subplot(1, 3, 3)
plt.plot(x_values, y_original, label='Fungsi Asli')
plt.plot(x_values, y_L4, label='L4 (Interpolasi Lagrange L4)')
plt.scatter(x_samples, y_samples, color='red', label='Titik Sampel')
plt.title('Interpolasi Polinomial Lagrange L4')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Nomor 1b

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Fungsi ketinggian terhadap waktu
def y(t):
    g = 9.81
    return 5 * t - 0.5 * g * t**2

# Input Data waktu
t_data = np.linspace(0, 2, 10)
y_data = y(t_data)

# Input Data waktu untuk plotting
t_plot = np.linspace(0, 2, 100)

# Fungsi-fungsi untuk fitting kurva
def linear_fit(t, a, b):
    return a * t + b

def quadratic_fit(t, a, b, c):
    return a * t**2 + b * t + c

def cubic_fit(t, a, b, c, d):
    return a * t**3 + b * t**2 + c * t + d

# Fitting kurva linier, kuadratik, dan kubik
popt_linear, _ = curve_fit(linear_fit, t_data, y_data)
popt_quadratic, _ = curve_fit(quadratic_fit, t_data, y_data)
popt_cubic, _ = curve_fit(cubic_fit, t_data, y_data)

# Plot data mentah dan fitting kurva
plt.figure(figsize=(10, 6))
plt.plot(t_data, y_data, 'bo', label='Data Mentah')

# Plot fitting kurva
plt.plot(t_plot, linear_fit(t_plot, *popt_linear), 'r--', label='Linier')
plt.plot(t_plot, quadratic_fit(t_plot, *popt_quadratic), 'g-', label='Kuadratik')
plt.plot(t_plot, cubic_fit(t_plot, *popt_cubic), 'm-', label='Kubik')

plt.title('Fitting Kurva Terhadap Data Mentah')
plt.xlabel('Waktu (s)')
plt.ylabel('Ketinggian (m)')
plt.grid(True)
plt.legend()
plt.show()

# Nomor 1c

# Hitung total kuadrat error
def total_squared_error(data, fit_func, params):
    y_fit = fit_func(data[0], *params)
    return np.sum((data[1] - y_fit)**2)

error_linear = total_squared_error((t_data, y_data), linear_fit, popt_linear)
error_quadratic = total_squared_error((t_data, y_data), quadratic_fit, popt_quadratic)
error_cubic = total_squared_error((t_data, y_data), cubic_fit, popt_cubic)

print('Total Kuadrat Error:')
print('Linier:', error_linear)
print('Kuadratik:', error_quadratic)
print('Kubik:', error_cubic)
```

```python
# Nomor 2

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Data jam dan temperatur
jam = np.array([6, 9, 12, 15, 18, 20])
temperatur = np.array([15, 18, 25, 22, 20, 19])

# Fungsi-fungsi untuk fitting kurva
def linear_fit(x, a, b):
    return a * x + b

def quadratic_fit(x, a, b, c):
    return a * x**2 + b * x + c

def cubic_fit(x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d

# Fitting kurva linier, kuadratik, dan kubik
popt_linear, _ = curve_fit(linear_fit, jam, temperatur)
popt_quadratic, _ = curve_fit(quadratic_fit, jam, temperatur)
popt_cubic, _ = curve_fit(cubic_fit, jam, temperatur)

# Plot data dan fitting kurva
plt.figure(figsize=(10, 6))
plt.plot(jam, temperatur, 'ro', label='Data')

# Plot fitting kurva
x_fit = np.linspace(min(jam), max(jam), 100)
plt.plot(x_fit, linear_fit(x_fit, *popt_linear), 'k--', label='Linier')
plt.plot(x_fit, quadratic_fit(x_fit, *popt_quadratic), 'b-', label='Kuadratik')
plt.plot(x_fit, cubic_fit(x_fit, *popt_cubic), 'c-', label='Kubik')

plt.title('Fitting Kurva terhadap Data Temperature')
plt.xlabel('Jam (Waktu)')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.legend()
plt.show()

# Hitung total kuadrat error
def total_squared_error(data, fit_func, params):
    x_data, y_data = data
    y_fit = fit_func(x_data, *params)
    return np.sum((y_data - y_fit)**2)

error_linear = total_squared_error((jam, temperatur), linear_fit, popt_linear)
error_quadratic = total_squared_error((jam, temperatur), quadratic_fit, popt_quadratic)
error_cubic = total_squared_error((jam, temperatur), cubic_fit, popt_cubic)

print('Total Kuadrat Error:')
print('Linier:', error_linear)
print('Kuadratik:', error_quadratic)
print('Kubik:', error_cubic)
```

Dijelaskan juga bagaimana kurva linier, kuadratik, dan kubik dipasang. Setiap fungsi ini akan digunakan untuk mensimulasikan hubungan antara jam dan suhu.
```python
import numpy as np
import matplotlib.pyplot as plt

# Data jam dan temperatur
jam_data = np.array([6, 9, 12, 15, 18, 20])
temp_data = np.array([15, 18, 25, 22, 20, 19])

# Derajat polinomial ekstrapolasi
degrees = [1, 2, 3, 4]

# Membuat data jam dari 0 hingga 24
jam_ekstrapolasi = np.linspace(0, 24, 100)

# Plot data asli
plt.figure(figsize=(10, 6))
plt.plot(jam_data, temp_data, 'bo', label='Data Asli')

# Ekstrapolasi data menggunakan polinomial dengan derajat berbeda
for degree in degrees:
    coeffs = np.polyfit(jam_data, temp_data, degree)
    temp_ekstrapolasi = np.polyval(coeffs, jam_ekstrapolasi)
    plt.plot(jam_ekstrapolasi, temp_ekstrapolasi, label=f'Polinomial Derajat {degree}')

plt.title('Ekstrapolasi Data Suhu')
plt.xlabel('Jam (24-jam)')
plt.ylabel('Temperatur (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data suhu
jam = np.array([6, 9, 12, 15, 18, 20])
suhu = np.array([15, 18, 25, 22, 20, 19])

# Menambahkan titik ekstrapolasi
jam_ekstrapolasi = np.linspace(0, 24, 100)
suhu_ekstrapolasi = np.zeros_like(jam_ekstrapolasi)
suhu_ekstrapolasi[:25] = np.linspace(-7, 14, 25)

# Fungsi untuk ekstrapolasi
def ekstrapolasi(x, y, derajat):
    coef = np.polyfit(x, y, derajat)
    return np.polyval(coef, jam_ekstrapolasi)

# Derajat ekstrapolasi
derajat_ekstrapolasi = [0, 1, 2, 3]

# Plotting
plt.figure(figsize=(10, 6))

# Plot data asli
plt.scatter(jam, suhu, color='brown', label='Data Asli')

# Ekstrapolasi dan plot untuk setiap derajat
for derajat in derajat_ekstrapolasi:
    ekstra_suhu = ekstrapolasi(jam, suhu, derajat)
    plt.plot(jam_ekstrapolasi, ekstra_suhu, label=f'Derajat {derajat}')

# Label dan judul
plt.xlabel('Waktu (Jam)')
plt.ylabel('Suhu (°C)')
plt.title('Ekstrapolasi Suhu Harian')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
