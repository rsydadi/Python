Interpolasi Polynomial
Dasar dari interpolasi polinomial melibatkan pencarian suatu polinomial yang melalui sejumlah titik data yang diberikan. Tujuan utamanya adalah untuk memperkirakan nilai di antara titik-titik data yang diketahui. Untuk memahami dasar interpolasi polinomial, pertama-tama kita perlu memahami konsep polinomial itu sendiri.

Sebuah polinomial adalah ekspresi matematika yang terdiri dari satu atau lebih suku, di mana setiap suku memiliki pangkat yang non-negatif dan koefisien yang mengalikan pangkat 
tersebut. Secara umum, polinomial n derajat memiliki bentuk seperti ini:
P(x) = a0 + a1x+a2x^2+...+anx^n

di mana a0,a1,...,an adalah koefisien polinomial, dan x adalah variabel independen.

Interpolasi polinomial bertujuan untuk menemukan polinomial P(x) dengan derajat tertentu n yang melalui sekumpulan titik data (x0,y0), (x1,y1), ..., (xm,ym). Dengan kata lain interpolasi polinomial berusaha untuk menemukan koefisien a0, a1, ..., an sehingga polinomial tersebut melewati semua titik data yang diberikan.

```python
import numpy as np
import matplotlib.pyplot as plt

# Titik data
x = np.array([0.5, 2.0, 4.0, 5.0, 7.0, 9.0])
y = np.array([0.5, 0.4, 0.3, 0.1, 0.9, 0.8])

# Derajat polinomial
degree = 5

# Menyesuaikan polinomial pada titik data
coefficients = np.polyfit(x, y, degree)

# Membuat fungsi polinomial dari koefisien
polynomial = np.poly1d(coefficients)

# Menampilkan koefisien polinomial
print("Koefisien polinomial:", coefficients)

# Membuat nilai x baru untuk evaluasi polinomial
x_new = np.linspace(min(x), max(x), 100)
y_new = polynomial(x_new)

# Plotting data asli dan polinomial hasil interpolasi
plt.scatter(x, y, color='red', label='Titik Data Asli')
plt.plot(x_new, y_new, label='Polinomial Interpolasi')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Interpolasi Polinomial Derajat {degree}')
plt.legend()
plt.show()
```

Untuk melakukan interpolasi polinomial pada sekumpulan titik data, langkah pertama adalah menciptakan array x dan y yang berisi titik-titik data yang dimiliki. Selanjutnya, cari derajat polinomial yang akan digunakan untuk interpolasi; kompleksitas polinomial yang akan digunakan ditentukan oleh derajat polinomial ini. Langkah berikutnya adalah menggunakan fungsi numpy.polyfit untuk mendapatkan koefisien polinomial yang sesuai dengan titik data. Fungsi ini akan menghasilkan array koefisien yang dapat digunakan untuk membentuk polinomial.

Untuk membuat fungsi polinomial dari koefisien yang telah diperoleh, gunakan numpy.poly1d. Fungsi ini memungkinkan evaluasi polinomial pada berbagai nilai x. Untuk melakukannya, kita membuat nilai-nilai baru dari x dalam rentang titik data asli, dan kemudian kita menggunakan fungsi polinomial yang telah kita buat untuk menghitung nilai y yang sesuai.

Langkah terakhir adalah memvisualisasikan hasil interpolasi dengan plotting menggunakan matplotlib.pyplot untuk menampilkan plot dari titik data asli dan kurva polinomial hasil interpolasi. Plot ini membantu kita memahami bagaimana polinomial yang dihasilkan sesuai dengan titik-titik data yang ada, dan bagaimana polinomial tersebut dapat digunakan untuk memperkirakan nilai di antara titik-titik data tersebut.
