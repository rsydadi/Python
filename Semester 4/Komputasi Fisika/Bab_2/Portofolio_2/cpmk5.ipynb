x_0 < x_1 < \cdots < x_n

Fitting kurva polinomial

Dalam konteks penggambaran data grafis, ketika diberikan sejumlah titik, tujuan utama adalah menemukan sebuah kurva matematika yang melalui semua titik tersebut. Dalam hal ini, kurva tersebut dapat direpresentasikan oleh sebuah fungsi matematika. Penting untuk diingat bahwa garis lurus merupakan kasus khusus dari sebuah kurva. Salah satu jenis fungsi matematika yang dapat digunakan untuk merepresentasikan kurva adalah fungsi polinomial.

Fungsi polinomial derajat satu pada dasarnya adalah garis lurus dan dinyatakan sebagai:
$f(x) = mx + c$

Fungsi polinomial derajat dua pada dasarnya adalah kurva dengan satu lekukan, dan dinyatakan sebagai:
$f(x) = ax^2 + bx + c$

Demikian pula, fungsi polinomial derajat $n$ adalah kurva dengan $n - 1$ lekukan.


```python
import numpy as np 
import matplotlib.pyplot as plt

# Perhatikan kembali contoh data di atas 
x = np.array([0.5, 2.0, 4.0, 5.0, 7.0, 9.0]) 
y = np.array([0.5, 0.4, 0.3, 0.1, 0.9, 0.8]) 
# contoh data fitting polinomial 
# membuat koefisien polyfit 
koef_1 = np.polyfit(x, y, 1)  # data fitting linier 
koef_2 = np.polyfit(x, y, 2)  # data fitting kuadrat f(x)=...x^2 + ...x+ ... 
koef_3 = np.polyfit(x, y, 3)  # data fitting kubik f(x)=....x^3 + ...x^2 + ...x+.. 
koef_4 = np.polyfit(x, y, 4)  # data fitting kubik f(x)=....x^4 + ...x^3 + ...x^2+.. 
koef_5 = np.polyfit(x, y, 5) 
#koef_6 = np.polyfit(x, y, 6) 
# membuat fungsi polinomial dari koefisien 
p1 = np.poly1d(koef_1) 
p2 = np.poly1d(koef_2) 
p3 = np.poly1d(koef_3) 
p4 = np.poly1d(koef_4) 
p5 = np.poly1d(koef_5) 
#p6 = np.poly1d(koef_6) 
# plot 
x_range = np.linspace(min(x), max(x), 500) # membuat titik x untuk plot data fitting 
# dari nilai x terendah sampai x terbesar sebanyak 500 titik 
plt.plot(x,y, 'kH', label='data mentah')  # plot data mentah 
plt.plot(x_range, p1(x_range), 'r-', label='data fitting orde 1') 
plt.plot(x_range, p2(x_range), 'b--', label='data fitting orde 2') 
plt.plot(x_range, p3(x_range), 'g-', label='data fitting orde 3') 
plt.plot(x_range, p4(x_range), 'm-', label='data fitting orde 4') 
plt.plot(x_range, p5(x_range), 'r:', label='data fitting orde 5') 
#plt.plot(x_range, p6(x_range), 'k--', label='data fitting orde 6')
plt.title('Aproksimasi polinomial dengan derajat yang berbeda', fontsize=16) 
plt.legend() 
plt.show() 
```

Contoh program ini menunjukkan implementasi fitting data polinomial menggunakan Python, yang dilakukan dengan menggunakan library NumPy dan matplotlib. Pertama, array numpy "x" dan "y" dibuat untuk data mentah yang akan difitting. Selanjutnya, program menggunakan metode "np.polyfit" untuk koefisien fitting polinomial untuk derajat polinomial 1 hingga 6.

Kemudian, fungsi polinomial dari koefisien yang dihasilkan dibuat menggunakan "np.poly1d". Selanjutnya, titik x baru untuk plot dibuat menggunakan "np.linspace" dari nilai x terendah hingga nilai x tertinggi dengan 500 titik.

Program kemudian memplot data mentah sebagai titik hitam (kH) dan kurva fitting polinomial untuk setiap derajat (orde 1 hingga 6) dengan berbagai warna dan gaya garis. Ini dilakukan dengan menggunakan "plt.plot". Dengan menggunakan "plt.legend" dan "plt.title", label dan judul dapat ditambahkan ke plot.

Hasil plot menunjukkan kurva fitting polinomial dengan berbagai derajat yang mengaproksimasi data mentah. Ini memberikan gambaran tentang bagaimana kurva fitting berubah sesuai dengan derajat polinomial yang digunakan untuk mengaproksimasi data.
