```python
import numpy as np 
import matplotlib.pyplot as plt 

# Perhatikan kembali contoh data di atas 
xi = np.array([0.5, 2.0, 4.0, 5.0, 7.0, 9.0]) 
yi = np.array([0.5, 0.4, 0.3, 0.1, 0.9, 0.8]) 

def hitung_coeff_newton_poly(xi, yi): 
    """ Evaluasi koefisien a_i secara rekursif menggunakan metode Newton 
    """ 
    # Inisialisasi array 'a' dengan yi, tetapi ambil salinannya untuk memastikan tidak 
    # menimpa data yi  
    a = yi.copy() 
    # Kita memiliki titik data N+1, sehingga 
    N = len(a) - 1 
    # untuk setiap k, kita menghitung Δ^k y_i dari a_i = Δ^(k-1) y_i dari iterasi 
    # sebelumnya: 
    for k in range(1, N+1): 
        # but only for i>=k 
        for i in range(k, N+1): 
            a[i] = (a[i] - a[k-1])/(xi[i]-xi[k-1]) 
    return a 

# dengan koefisien a, dan lokasi data xi, 
# menentukan fungsi untuk mengevaluasi polinomial Newton 
# di lokasi yang diberikan dalam array x. 
# NB. ini hanya evaluasi persamaan P_n(x) = ... 
# yang diberikan di awal bagian ini. 
def eval_newton_poly(a, xi, x): 
    """ Berfungsi untuk mengevaluasi polinomial Newton 
    di x, diketahui titik data xi dan koefisien polinomial a 
    """ 
    N = len(xi) - 1  # derajat polinomial 
    # secara rekursif membangun polinomial yang dievaluasi di x 
    P = a[N] 
    for k in range(1, N+1): 
        P = a[N-k] + (x - xi[N-k])*P 
    return P

# set up grafik 
fig = plt.figure(figsize=(7, 5)) 
ax1 = fig.add_subplot(111) 
# tambahkan margin kecil  
ax1.margins(0.1) 
# Evaluasi koefisien polinomial Newton 
a = hitung_coeff_newton_poly(xi, yi) 
# Evaluasi polinomial pada resolusi tinggi dan plot  
x = np.linspace(0.4, 9.1, 100) 
ax1.plot(x, eval_newton_poly(a, xi, x), 'b', label='Newton polynomial') 
# menimpa data mentah 
plot_data_mentah = (xi, yi, ax1) 
ax1.set_title('Interpolasi polinomial Newton', fontsize=16) 
# menambahkan legend 
ax1.legend(loc='best', fontsize=14) 
plt.xlim(0,9.2) 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data mentah
x = np.array([0.5, 2.0, 4.0, 5.0, 7.0, 9.0])
y = np.array([0.5, 0.4, 0.3, 0.1, 0.9, 0.8])

# Cocokkan polinomial derajat 1 (linear fit)
degree = 1
poly_coeffs = np.polyfit(x, y, degree)
print('poly_coeffs: ', poly_coeffs)

# Gunakan poly1d untuk mengubah koefisien menjadi fungsi, p1, kita dapat mengevaluasi
p1 = np.poly1d(poly_coeffs)

# Plotting
x_range = np.linspace(min(x), max(x), 500)
plt.figure(figsize=(8, 5))
plt.plot(x_range, p1(x_range), label='Fitting polinomial', color='blue')
plt.scatter(x, y, color='red', label='Data mentah')
plt.title('Fitting polinomial ke Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.xlim(0, 9.2)
plt.show()
```

```python
# Perhatikan kembali contoh data di atas 
x = np.array([0.5, 2.0, 4.0, 5.0, 7.0, 9.0]) 
y = np.array([0.5, 0.4, 0.3, 0.1, 0.9, 0.8]) 
# contoh data fitting polinomial 
# membuat koefisien polyfit 
koef_1 = np.polyfit(x, y, 1)  # data fitting linier 
koef_2 = np.polyfit(x, y, 2)  # data fitting kuadrat f(x)=...x^2 + ...x+ ... 
koef_3 = np.polyfit(x, y, 3)  # data fitting kubik f(x)=....x^3 + ...x^2 + ...x+.. 
koef_4 = np.polyfit(x, y, 4)  # data fitting kubik f(x)=....x^4 + ...x^3 + ...x^2+.. 
koef_5 = np.polyfit(x, y, 5) 
koef_6 = np.polyfit(x, y, 6) 
# membuat fungsi polinomial dari koefisien 
p1 = np.poly1d(koef_1) 
p2 = np.poly1d(koef_2) 
p3 = np.poly1d(koef_3) 
p4 = np.poly1d(koef_4) 
p5 = np.poly1d(koef_5) 
p6 = np.poly1d(koef_6) 
# plot 
x_range = np.linspace(min(x), max(x), 500) # membuat titik x untuk plot data fitting 
# dari nilai x terendah sampai x terbesar sebanyak 500 titik 
plt.plot(x,y, 'kH', label='data mentah')  # plot data mentah 
plt.plot(x_range, p1(x_range), 'r-', label='data fitting orde 1') 
plt.plot(x_range, p2(x_range), 'b--', label='data fitting orde 2') 
plt.plot(x_range, p3(x_range), 'g-', label='data fitting orde 3') 
plt.plot(x_range, p4(x_range), 'm-', label='data fitting orde 4') 
plt.plot(x_range, p5(x_range), 'r:', label='data fitting orde 5') 
plt.plot(x_range, p6(x_range), 'k--', label='data fitting orde 6')
plt.title('Aproksimasi polinomial dengan derajat yang berbeda', fontsize=16) 
plt.legend() 
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi ketinggian terhadap waktu
def y(t):
    g = 9.81
    return 5*t - 0.5*g*t**2

# Input Waktu
t = np.linspace(0, 2, 100)

# Ketinggian bola terhadap waktu
height = y(t)

# Plot grafik
plt.plot(t, height, "r")
plt.title('Pergerakan Bola')
plt.xlabel('Waktu (s)')
plt.ylabel('Ketinggian (m)')
plt.grid(True)
plt.show()
```
