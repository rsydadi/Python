```python
def RungeKutta(x, y, dx, dydx):
    # Menghitung kemiringan
    k1 = dx * dydx(x, y)
    k2 = dx * dydx(x + dx/2, y + k1/2)
    k3 = dx * dydx(x + dx/2, y + k2/2)
    k4 = dx * dydx(x + dx, y + k3)
    
    # Menghitung nilay x dan y yang baru
    y = y + 1./6*(k1 + 2*k2 + 2*k3 + k4)
    x = x + dx
    return x, y

def dydx(x, y):
    return x + y

def y1(x):
    return x**3/3.+2/3

x0 = 1
y0 = 1
dx = 0.1
x_end = 2

x_rk = [x0]
y_rk = [y0]

y = y0
x = x0
while x < x_end:
    x, y = RungeKutta(x, y, dx, dydx)
    x_rk.append(x)
    y_rk.append(y)

def Euler(x, y, dx, dydx):
    return x+dx, y+dx*dydx(x, y)

x_eu= [x0]
y_eu = [y0]

while x < x_end:
    x, y = Euler(x, y, dx, dydx)
    x_eu.append(x)
    y_eu.append(y)


import numpy as np 
import matplotlib.pyplot as plt 
 
plt.figure(figsize=(7, 5))

plt.plot(np.linspace(1, 2, 50), y1(np.linspace(1, 2, 50)), label= 'Analitik')
plt.plot(x_rk, y_rk, 'r--o', label="Solusi Numerik:\nRunge-Kutta") 
plt.plot(x_eu, y_eu, 'b--o', label="Solusi Numerik:\nEuler") 
plt.legend(loc="best", fontsize=12)
plt.title(r"Solusi PDB: $\quad\frac{dy}{dx}=x^2$") 
plt.xlabel("x", fontsize=12) 
plt.ylabel("y", fontsize=12) 
plt.show()
```
