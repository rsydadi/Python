```python
# Nomor 2

import matplotlib.pyplot as plt
import numpy as np

def euler(f, u0, t0, t_max, dt): 
    u = u0
    t = t0 
    u_all = [u0]
    t_all = [t0]
    while t < t_max: 
        u = u + dt * f(u, t) 
        u_all.append(u)
        t = t + dt 
        t_all.append(t)
    return u_all, t_all 

def f1(u, t): 
    return u

def f2(u, t): 
    return np.cos(t)

# Definisikan berbagai langkah waktu
dts = [0.1, 0.01, 0.001]

# Solusi eksak
def solusi_eksak1(t):
    return np.exp(t)

def solusi_eksak2(t):
    return np.sin(t) + 1

# Plot solusi numerik dan solusi eksak untuk berbagai langkah waktu
fig, axes = plt.subplots(3, 2, figsize=(12, 12))
for i, dt in enumerate(dts):
    u1, t1 = euler(f1, 1.0, 0.0, 10.0, dt)
    u2, t2 = euler(f2, 0.0, 0.0, 10.0, dt)

    ax1 = axes[i, 0]
    ax2 = axes[i, 1]
    
    # Plot untuk fungsi pertama
    ax1.plot(t1, solusi_eksak1(np.array(t1)), color="red", label="Eksak $u(t)=\exp(t)$")
    ax1.plot(t1, u1, color="blue", label=f"Numerik $u'(t)=u(t)$,\n$u(0)=1$, $\\Delta t={dt}$")
    ax1.set_xlabel('$t$', fontsize=14)
    ax1.set_ylabel('$u(t)$', fontsize=14)
    ax1.legend(loc="best", fontsize=12)
    
    # Plot untuk fungsi kedua
    ax2.plot(t2, solusi_eksak2(np.array(t2)), color="red", label="Eksak $u(t)=\sin(t)+1$")
    ax2.plot(t2, u2, color="blue", label=f"Numerik $u'(t)=\cos(t)$,\n$u(0)=0$, $\\Delta t={dt}$")
    ax2.set_xlabel('$t$', fontsize=14)
    ax2.set_ylabel('$u(t)$', fontsize=14)
    ax2.legend(loc="lower left", fontsize=12)

plt.subplots_adjust(wspace=0.3, hspace=0.5)
plt.suptitle("Solusi dengan Metode Euler untuk Berbagai $\Delta t$", fontsize=16)
plt.show()
```

```python
# Nomor 1

import matplotlib.pyplot as plt
import numpy as np

def RungeKutta(x, y, dx, dydx):
    # menghitung kemiringan 
    k1 = dx * dydx(x, y) 
    k2 = dx * dydx(x + dx / 2., y + k1 / 2.) 
    k3 = dx * dydx(x + dx / 2., y + k2 / 2.) 
    k4 = dx * dydx(x + dx, y + k3) 

    # menghitung x dan y yang baru 
    y = y + 1. / 6 * (k1 + 2 * k2 + 2 * k3 + k4) 
    x = x + dx 
    return x, y 

def dydx1(x, y): 
    return 1 / (x * y)

def y1(x): 
    return np.sqrt(1 + 2 * np.log(x))

x0 = 1. 
y0 = 1. 
dx = 0.1 
x_end = 2
x_rk = [x0] 
y_rk = [y0] 
y = y0 
x = x0 

while x < x_end: 
    x, y = RungeKutta(x, y, dx, dydx1) 
    x_rk.append(x) 
    y_rk.append(y) 

def Euler(x, y, dx, dydx): 
    return x + dx, y + dx * dydx(x, y) 

x_eu = [x0] 
y_eu = [y0] 
y = y0 
x = x0 

while x < x_end: 
    x, y = Euler(x, y, dx, dydx1) 
    x_eu.append(x) 
    y_eu.append(y) 

plt.figure(figsize=(7,5)) 
plt.plot(np.linspace(1, 2, 50), y1(np.linspace(1, 2, 50)), label="Solusi Analitis", color="black")
plt.plot(x_rk, y_rk, 'r--o', label="Solusi Numerik:\nRunge-Kutta") 
plt.plot(x_eu, y_eu, 'b--o', label="Solusi Numerik:\nEuler") 
plt.legend(loc="best", fontsize=12) 
plt.title(r"Solusi PDB: $\quad\frac{dy}{dx}=\frac{1}{xy}$") 
plt.xlabel("x", fontsize=12) 
plt.ylabel("y", fontsize=12) 
plt.show()
```

```python
# Nomor 1 dan  2

import matplotlib.pyplot as plt
import numpy as np

# Define parameters
omega = np.pi
gamma = 0.1
t0, t_end, dt = 0, 20, 0.01
N = int((t_end - t0) / dt) + 1
t = np.linspace(t0, t_end, N)

# Initialize arrays for x and v
x1, v1 = np.zeros(N), np.zeros(N)
x2, v2 = np.zeros(N), np.zeros(N)
x1[0], v1[0] = 1, 0
x2[0], v2[0] = 1, 0

# Runge-Kutta method for simple harmonic oscillator
for i in range(N - 1):
    k1_x1 = dt * v1[i]
    k1_v1 = dt * (-omega**2 * x1[i])
    k2_x1 = dt * (v1[i] + 0.5 * k1_v1)
    k2_v1 = dt * (-omega**2 * (x1[i] + 0.5 * k1_x1))
    k3_x1 = dt * (v1[i] + 0.5 * k2_v1)
    k3_v1 = dt * (-omega**2 * (x1[i] + 0.5 * k2_x1))
    k4_x1 = dt * (v1[i] + k3_v1)
    k4_v1 = dt * (-omega**2 * (x1[i] + k3_x1))
    x1[i + 1] = x1[i] + (k1_x1 + 2 * k2_x1 + 2 * k3_x1 + k4_x1) / 6
    v1[i + 1] = v1[i] + (k1_v1 + 2 * k2_v1 + 2 * k3_v1 + k4_v1) / 6

# Runge-Kutta method for damped harmonic oscillator
for i in range(N - 1):
    k1_x2 = dt * v2[i]
    k1_v2 = dt * (-omega**2 * x2[i] - 2 * gamma * v2[i])
    k2_x2 = dt * (v2[i] + 0.5 * k1_v2)
    k2_v2 = dt * (-omega**2 * (x2[i] + 0.5 * k1_x2) - 2 * gamma * (v2[i] + 0.5 * k1_v2))
    k3_x2 = dt * (v2[i] + 0.5 * k2_v2)
    k3_v2 = dt * (-omega**2 * (x2[i] + 0.5 * k2_x2) - 2 * gamma * (v2[i] + 0.5 * k2_v2))
    k4_x2 = dt * (v2[i] + k3_v2)
    k4_v2 = dt * (-omega**2 * (x2[i] + k3_x2) - 2 * gamma * (v2[i] + k3_v2))
    x2[i + 1] = x2[i] + (k1_x2 + 2 * k2_x2 + 2 * k3_x2 + k4_x2) / 6
    v2[i + 1] = v2[i] + (k1_v2 + 2 * k2_v2 + 2 * k3_v2 + k4_v2) / 6

# Plot the results
plt.figure(figsize=(12, 6))

# Simple Harmonic Oscillator
plt.subplot(2, 1, 1)
plt.plot(t, x1, 'r-', label='Perpindahan')
plt.plot(t, v1, 'b--', label='Kecepatan')
plt.title('Osilator Harmonik Sederhana')
plt.xlabel('Waktu')
plt.ylabel('Amplitudo')
plt.legend()
plt.grid(True)

# Damped Harmonic Oscillator
plt.subplot(2, 1, 2)
plt.plot(t, x2, 'r-', label='Perpindahan')
plt.plot(t, v2, 'b--', label='Kecepatan')
plt.title('Osilator Harmonik Teredam')
plt.xlabel('Waktu')
plt.ylabel('Amplitudo')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Dengan demikian, program ini membantu memahami bagaimana parameter redaman memengaruhi perilaku sistem beban-pegas dalam konteks simulasi numerik.
